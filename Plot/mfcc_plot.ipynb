```python
import os
import matplotlib.pyplot as plt
from python_speech_features import mfcc
import scipy.io.wavfile as wav
import numpy as np
from scipy.signal import butter,lfilter,freqz
```

```python
def compute_mfcc(folder_path):
    mfcc_features = []
    labels = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.wav'):
            (rate, sig) = wav.read(os.path.join(folder_path, filename))
            mfcc_feat = mfcc(sig, rate, numcep=12, nfilt=26, nfft=2048)
            mfcc_features.append(mfcc_feat)
            labels.append(folder_path.split('/')[-1])
    return mfcc_features, labels
```

```python

noise_mfcc, noise_labels = compute_mfcc('D:\\HK2-Năm 3\\PBL5\\Code\\data\\Noise')
crying_mfcc, crying_labels = compute_mfcc('D:\\HK2-Năm 3\\PBL5\\Code\\data\\Crying')
laugh_mfcc, laugh_labels = compute_mfcc('D:\\HK2-Năm 3\\PBL5\\Code\\data\\Laugh')
silence_mfcc, silence_labels = compute_mfcc('D:\\HK2-Năm 3\\PBL5\\Code\\data\\Silence')

# Tạo biểu đồ cho đặc trưng MFCC của mỗi thư mục
fig, axs = plt.subplots(4, 1, figsize=(10, 10))
fig.tight_layout(pad=3.0)
axs[0].set_title('Noise')
for i, mfcc_feat in enumerate(noise_mfcc):
    axs[0].imshow(mfcc_feat.T, origin='lower', aspect='auto', cmap='jet', interpolation='nearest')
    axs[0].set_ylabel('MFCC coefficients')
    axs[0].set_xlabel('Frame')
    axs[0].set_yticks(np.arange(0, mfcc_feat.shape[1], 2))
    axs[0].set_yticklabels(np.arange(0, mfcc_feat.shape[1], 2))
    axs[0].grid(False)
axs[1].set_title('Crying')
for i, mfcc_feat in enumerate(crying_mfcc):
    axs[1].imshow(mfcc_feat.T, origin='lower', aspect='auto', cmap='jet', interpolation='nearest')
    axs[1].set_ylabel('MFCC coefficients')
    axs[1].set_xlabel('Frame')
    axs[1].set_yticks(np.arange(0, mfcc_feat.shape[1], 2))
    axs[1].set_yticklabels(np.arange(0, mfcc_feat.shape[1], 2))
    axs[1].grid(False)
    axs[2].set_title('Laugh')
for i, mfcc_feat in enumerate(laugh_mfcc):
    axs[2].imshow(mfcc_feat.T, origin='lower', aspect='auto', cmap='jet', interpolation='nearest')
    axs[2].set_ylabel('MFCC coefficients')
    axs[2].set_xlabel('Frame')
    axs[2].set_yticks(np.arange(0, mfcc_feat.shape[1], 2))
    axs[2].set_yticklabels(np.arange(0, mfcc_feat.shape[1], 2))
    axs[2].grid(False)
axs[3].set_title('Silence')
for i, mfcc_feat in enumerate(silence_mfcc):
    axs[3].imshow(mfcc_feat.T, origin='lower', aspect='auto', cmap='jet', interpolation='nearest')
    axs[3].set_ylabel('MFCC coefficients')
    axs[3].set_xlabel('Frame')
    axs[3].set_yticks(np.arange(0, mfcc_feat.shape[1], 2))
    axs[3].set_yticklabels(np.arange(0, mfcc_feat.shape[1], 2))
    axs[3].grid(False)

plt.show()
```
